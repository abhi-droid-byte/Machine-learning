```python
plt.plot(x,y,'r*--')
plt.show()
plt.plot(x,y,'b',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph 2')
plt.show()
```

```python
#creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.subplot(2,2,3)
plt.plot(x,y,'r')
```

```python

#Compute the x and y coordinates for points on sine and cosine curves
x= np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

#Set up a subplot grid that has height 2 and width 1,
#and set the first such subplot as active.

plt.subplot(2, 1, 1)

#Make the first plot
plt.plot(x, y_sin,'r--')
plt.title('Sine')


#
plt.subplot(2, 1, 2)
plt.plot(x,y_cos,'g--')
plt.title('Cosine')


plt.show()
```

```python
#Bar Plot
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6, 15 , 7]

plt.bar(x, y, align = 'center')
plt.bar(x2, y2, color = 'g')
plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y  axis')
plt.show()
```

```python
a= np.array([22,87,5,56,73,55,11,20,51,5,40,40,79,31,27])
plt.hist(a)
plt.title("histogram")
plt.show()
```

```python
#box plots helps you to find percentile

data= [np.random.normal(0, std, 100) for std in range(1,4)]

#rectangle box plot

plt.boxplot(data, vert=True, patch_artist=False)
```

```python
#pie chart
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  #explode 1st slice


#plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct= '%1.1f%%', shadow= True)

plt.axis('equal')
plt.show()
```
